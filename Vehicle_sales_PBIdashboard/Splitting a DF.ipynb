```python
import pandas as pd
df = pd.read_csv("Car_prices_clean.csv")
```

```python
#Checking for nulls
df.isnull().any()
```

```python
df.head(1)
```

```python
#Creating a list of columns
df.columns
```

```python
#Creating a list of columns that will go to VIN data frame
columns_vin = ['VIN', 'Make', 'Model', 'Year', 'Trim', 'Body', 'Transmission', 'Color', 'Interior']
```

```python
#Creating VIN data frame
df_vin = pd.DataFrame(columns = columns_vin)
```

```python
#filling data frame with data
df_vin[columns_vin] = df[columns_vin]
```

```python
df_vin
```

```python
#creating a list to store DF column name
columns_seller = ['Seller']
```

```python
#Creating a data frame to store "Seller" names
df_seller = pd.DataFrame(columns = columns_seller)
```

```python
#selecting distinct Sellers from the main DF
df_seller["Seller"] = df["Seller"].unique()
```

```python
#Checking the output
df_seller
```

```python
#Creating column Seller ID and assigning a range of values to uniquely identify each row 
df_seller["Seller_ID"] = range(1, len(df_seller)+1)
```

```python
#Setting 'Seller' column as the index column and then convert DF into a dictionary. 
#In the dictionary, the keys will be Seller Names 
seller_dict = df_seller.set_index('Seller')['Seller_ID'].to_dict()
```

```python
#Now we are replacing the mathcing values (dictionary keys) in DF with the dictionary values
df['Seller'] = df['Seller'].map(seller_dict)
```

```python
df = df.rename(columns = {"Seller": "Seller_ID"})
```

```python
df
```

```python
#Creating a list of columns that will go to sales_conditions data frame
columns_sales_conditions=['VIN','Condition', 'Odometer']
```

```python
#creating data frame
df_sales_conditions = pd.DataFrame(columns = columns_sales_conditions)
```

```python
#filling data frame with data
df_sales_conditions[columns_sales_conditions] = df[columns_sales_conditions]
```

```python
df_sales_conditions
```

```python
#Creating a list of columns that will go to sales fact dataframe
columns_fact_sales = ['VIN', 'Seller_ID', 'State', 'MMR', 'Selling_Price', 'Sale_Year', 'Sale_Day', 'Sale_Month_And_Day', 'Sale_Time']
```

```python
df_fact_sales = pd.DataFrame(columns = columns_fact_sales)
```

```python
df_fact_sales
```

```python
#filling data frame with data
df_fact_sales[columns_fact_sales] = df[columns_fact_sales]
```

```python
#Adding numeric column representing quantity of vehicles sold
df_fact_sales["Quantity_Sold"] = 1
```

```python
df_fact_sales
```

```python
#querying a list of columns 
df_fact_sales.columns
```

```python
#reodering columns in a logical fashion
col_update = ['VIN', 'Seller_ID', 'Quantity_Sold', 'State', 'MMR', 'Selling_Price', 'Sale_Year',
       'Sale_Day', 'Sale_Month_And_Day', 'Sale_Time']
```

```python
#updating column names
df_fact_sales = df_fact_sales[col_update]
```

```python
df_fact_sales
```

```python
#finally saving data frames as comma separated values files
df_vin.to_csv('VIN_details.csv', index=False)
```
