```python
#Importing Pandas
import pandas as pd
```

```python
#Reading csv file
df = pd.read_csv("car_prices.csv")
```

```python
df
```

```python
#Uppecasing VIN-numbers:
df["vin"] = df["vin"].str.upper()
```

```python
#Capitalizing column names
df.columns = [col.capitalize() for col in df.columns]
df.columns
```

```python
df.head(5)
```

```python
#Renaming columns for better readability
df.rename(columns = {"Mmr": "MMR", "Sellingprice": "Selling_Price", "Saledate": "Sale_Date", "Vin": "VIN"}, inplace = True)
```

```python
#Uppercasing State
df["State"] = df["State"].str.upper()
```

```python
#capitalizing each word in Seller column:
df["Seller"] = df["Seller"].str.title()
```

```python
df
```

```python
# Apply the title() method to each cell in the specified columns
df["Make"] = df["Make"].str.title()
df["Model"] = df["Model"].str.title()
df["Body"] = df["Body"].str.title()
df["Transmission"] = df["Transmission"].str.title()
df["Color"] = df["Color"].str.title()
df["Interior"] = df["Interior"].str.title()
```

```python
df["Body"] = df["Body"].str.replace('Suv', 'SUV')
```

```python
df["Make"] = df["Make"].str.upper()
```

```python
df
```

```python
#hecking VINs to identify NULLs and figure out what to do with them. 
#We see that 4 pcs of VINs were mistakenly inserted in the 'State' Column, so I will move them back into 'VIN' column
df[df["VIN"].isnull()]
```

```python
#After we've identified the indexes, we insert VINs into the 'VIN' column:
df.loc[461612, "VIN"] = "3VWD17AJ3FM259017"
df.loc[505299, "VIN"] = "3VWD17AJ7FM222388"
df.loc[529009, "VIN"] = "3VWD17AJ8FM298895"
df.loc[551222, "VIN"] = "3VWD17AJ8FM239622"
```

```python
#and remove VINs from the 'State' column:
df.loc[505299, "State"] = ""
df.loc[529009, "State"] = ""
df.loc[551222, "State"] = ""
df.loc[461612, "State"] = ""
```

```python
#Now removing rows where Sale_Date is not defined
df = df[df["Sale_Date"] != ""]
```

```python
#Let's locate rows where number of characters in the 'State' column is more than two, it will show us which the list of indices where State values should be fixed. 
df.loc[df['State'].str.len() > 2].index
```

```python
#We found that 22 VINS were mistakenly inserted into the 'State' column:
specific_indexes = [408161, 417835, 421289, 424161, 427040, 427043, 434424, 444501, 453794,
       461597, 492484, 497008, 497011, 499083, 501455, 505306, 520461, 528996,
       529013, 529622, 548784, 554710]
```

```python
#Let's copy them from 'States' to 'VINs'
df.loc[specific_indexes, "VIN"] = df.loc[specific_indexes, "State"]
```

```python
#And clean up the mess in the 'State' column
df.loc[specific_indexes, "State"] = ""
```

```python
#A typical Sale_Date column is comprised of 39 characters, let's locate those, which contains less than 39:
df[df["Sale_Date"].str.len() < 39]
df.loc[df["Sale_Date"].str.len() < 39, "Sale_Date"] = ""
```

```python
#Cleaning VINs containing / at the beginning
df["VIN"] = df["VIN"].str.lstrip("/") 
```

```python
df
```

```python
#Dropping nulls before splitting the column
df = df.dropna(subset = ["Sale_Date"])
```

```python
#adding new column with a split Sale_Date
df["New_Sale_Date"] = df["Sale_Date"].apply(lambda x: x[0:10]+ '/' + x[11:15] + '/' + x[16:])
```

```python
#Extracting Year
df["Sale_Year"] = df['New_Sale_Date'].str.extract(r'/([^/]*)/')
```

```python
#Extracting Day of Week
df['Sale_Day'] = df['New_Sale_Date'].str.slice(0,3)
```

```python
#Extracting Month and Day
df["Sale_Month_And_Day"] = df['New_Sale_Date'].str.slice(4,10)
```

```python
#Extracting time
df['Sale_Time'] = df['New_Sale_Date'].str.slice(17,) 
```

```python
#Deleting New_Sale_Date column
df = df.drop(["New_Sale_Date"], axis=1)
```

```python
df['Condition'] = pd.to_numeric(df['Condition'], errors='coerce')
df['Odometer'] = pd.to_numeric(df['Odometer'], errors='coerce')
df['MMR'] = pd.to_numeric(df['MMR'], errors='coerce')
df['Selling_Price'] = pd.to_numeric(df['Selling_Price'], errors='coerce').round()
```

```python
#df_copy[df_copy["Transmission"] == "Sedan"]
df["Transmission"] = df["Transmission"].str.replace('Sedan', '')
```

```python
#df_copy[df_copy["Color"]=="â€”"]
df["Color"] = df["Color"].str.replace('â€”', '')
```

```python
df['Color'] = df['Color'].str.replace('[0-9]', '', regex = True)
```

```python
df["Interior"] = df["Interior"].str.replace('â€”', '')
```

```python
df.columns
```

```python
#Defining the list of columns where nulls will be replaced with ""
columns_to_fill= {
"Make":              "",
"Model":            "",
"Trim":             "",
"Body":             "",
"Transmission":     "",
"Condition":        "",
"Odometer":         "",
"Color":            "",
"Interior":         "",
"MMR":              "",
"Selling_Price":    "",
}
```

```python
df.fillna(columns_to_fill, inplace=True)
```

```python
#Final check for nulls
df.isnull().any()
```
